```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv("data.csv")
data.drop(columns=["variety"],inplace=True)
```

```python
data1=data.to_numpy()
print(data1)
print(type(data1))
```

```python
kmeans_model=KMeans(n_clusters=3,random_state=1).fit(data1)
labels=kmeans_model.labels_
clusters_sklearn=kmeans_model.cluster_centers_
print(labels)
print(clusters_sklearn)
```

```python
plt.scatter(data1[:,0],data1[:,1],c=labels)
plt.scatter(clusters_sklearn[:,0],clusters_sklearn[:,1],marker="X",color="red")
```
