```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
```

```python
data=pd.read_csv('iris.csv')
```

```python
print(data)
```

```python
print(data.info())
```

```python
plt.plot(data["variety"])
plt.xlabel("Numbers")
plt.ylabel("variety")
plt.show()
```

```python
plt.hist(data["variety"])
plt.show()
```

```python
sb.scatterplot(data)
```

```python
sb.scatterplot(data["sepal.length"])
```

```python
sb.scatterplot(data,x="sepal.width",y="sepal.length",hue="petal.width")
```

```python
sb.pairplot(data,hue="variety")
plt.show()
```

```python
sb.FacetGrid(data,hue="variety").map(sb.histplot,"petal.width",kde=True).add_legend()
```

```python
sb.boxplot(data,x="variety",y="sepal.length",hue="variety")
```

```python
sb.violinplot(data,hue="variety",x="variety",y="sepal.width",linecolor="red")
plt.grid()
```

```python
import plotly.express as pt
```

```python
pt.scatter_3d(data,x="sepal.width",y="petal.length",z="petal.width",color="variety")
```
