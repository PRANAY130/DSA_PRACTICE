```python
iris=pd.read_csv("iris.csv")
iris.drop(columns=["variety"],inplace=True)
data=iris.to_numpy()
print(data)
```

```python
list1=[]
for i in range(1,11):
    model=KMeans(n_clusters=i,random_state=1).fit(data)
    labels=model.labels_
    cluster=model.cluster_centers_
    inertia=model.inertia_
    list1.append(inertia)
    plt.scatter(data[:,0],data[:,1],c=labels,label="Data Points")
    plt.scatter(cluster[:,0],cluster[:,1],marker="X",color="red",label="Centers")
    plt.grid()
    plt.legend()
    plt.show()
    
plt.plot(list1)
plt.grid()
plt.show()
```
